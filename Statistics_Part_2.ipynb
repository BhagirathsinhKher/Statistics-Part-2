Q1. What is hypothesis testing in statistics?

- Hypothesis testing is a statistical method used to determine if there is enough evidence to reject a given assumption (hypothesis) about a population parameter.

- **Steps in Hypothesis Testing :**
1. **Formulate Hypotheses**  
   - **Null Hypothesis (\(H_0\))**: Assumes no effect or no difference.  
   - **Alternative Hypothesis (\(H_1\))**: Represents what we want to test or prove.

2. **Choose a Significance Level (\(\alpha\))**  
   - Common values: 0.05 (5%) or 0.01 (1%).  
   - This represents the probability of rejecting \(H_0\) when it is actually true.

3. **Select a Test Statistic**  
   - Examples: Z-test, t-test, Chi-square test, ANOVA, etc.  
   - Depends on sample size and data type.

4. **Compute the Test Statistic**  
   - Compare the computed test statistic against the critical value.

5. **Make a Decision**  
   - Compare the **p-value** with \(\alpha\).  
   - If \(p \leq \alpha\), reject \(H_0\); otherwise, fail to reject \(H_0\).

Q2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?

- Null Hypothesis (\(H_0\)) vs. Alternative Hypothesis (\(H_1\))

1. What is the Null Hypothesis (\(H_0\))?

- The **null hypothesis (\(H_0\))** is the default assumption in hypothesis testing. It states that there is **no effect, no difference, or no relationship** between variables. It represents the **status quo** or an assumption that we seek to test.

- Example: A factory claims that the average weight of a product is 50 kg.  
  - **\(H_0\)**: The mean weight is **50 kg** (\(\mu = 50\)).

2. What is the Alternative Hypothesis (\(H_1\))?
- The **alternative hypothesis (\(H_1\))** is the opposite of the null hypothesis. It suggests that there **is an effect, a difference, or a relationship** between variables.

- Example (continuing from above):  
  - **\(H_1\)**: The mean weight is **not** 50 kg (\(\mu \neq 50\)).

3. Key Differences Between \(H_0\) and \(H_1\)

| Aspect              | Null Hypothesis (\(H_0\)) | Alternative Hypothesis (\(H_1\)) |
|---------------------|-----------------|---------------------|
| Meaning | Assumes no change or effect | Assumes a change or effect |
| Symbol | \(H_0\) | \(H_1\) |
| Example | Mean weight = 50 kg (\(\mu = 50\)) | Mean weight ≠ 50 kg (\(\mu \neq 50\)) |
| Aim | To test if it should be rejected | To support if evidence is strong |
| Decision | Either **reject** or **fail to reject** | If \(H_0\) is rejected, \(H_1\) is accepted |

Q3. What is the significance level in hypothesis testing, and why is it important?
1. What is the Significance Level (α)?  
The significance level (α) is the probability of rejecting the null hypothesis (H0) when it is actually true. It represents the risk of making a Type I error (false positive).  

Common values of α:  
- 0.05 (5%) → Most commonly used  
- 0.01 (1%) → Used for more stringent tests  
- 0.10 (10%) → Sometimes used in exploratory studies  

If α = 0.05, it means we are willing to accept a 5% chance of incorrectly rejecting H0.  

2. Why is the Significance Level Important?  
- Helps decide when to reject or fail to reject H0.  
- Controls the risk of Type I errors (false positives).  
- Balances between detecting real effects and avoiding false alarms.  

Types of errors in hypothesis testing:  

Type I Error: Rejecting H0 when it is true (controlled by α).  
Type II Error: Failing to reject H0 when it is false (controlled by test power, 1 - β).

Q4. What does a P-value represent in hypothesis testing?
1. What is a P-Value?  
- The p-value is the probability of observing the test results (or more extreme results) assuming that the null hypothesis (H0) is true. It helps determine whether the observed data is statistically significant.  

- A small p-value indicates that the observed data is unlikely under H0, leading to the rejection of H0. A large p-value suggests that the data is consistent with H0.  

2. Interpreting P-Value:  
- If **p-value ≤ α (significance level, e.g., 0.05)** → Reject H0 (evidence suggests a significant effect).  
- If **p-value > α** → Fail to reject H0 (not enough evidence to support an effect).

Q5. How do you interpret the P-value in hypothesis testing?
1. What is the P-Value?  
The p-value is the probability of obtaining test results at least as extreme as the observed data, assuming the null hypothesis (H0) is true. It helps determine whether there is enough evidence to reject H0.  

2. How to Interpret the P-Value?  
- **Small p-value (≤ α, e.g., 0.05)** → Strong evidence against H0 → **Reject H0**  
- **Large p-value (> α, e.g., 0.05)** → Weak evidence against H0 → **Fail to reject H0**  

Common significance levels (α):  
- 0.01 (1%) → Very strong evidence required to reject H0  
- 0.05 (5%) → Standard threshold  
- 0.10 (10%) → More lenient threshold

Q6.What are Type 1 and Type 2 errors in hypothesis testing?
1. What are Type I and Type II Errors?  
In hypothesis testing, we make a decision to either reject or fail to reject the null hypothesis (H0). However, errors can occur in this decision process.  

- **Type I Error (False Positive):** Rejecting H0 when it is actually true.  
- **Type II Error (False Negative):** Failing to reject H0 when it is actually false.  

2. Understanding Type I Error:  
- Occurs when we incorrectly conclude that there is an effect when none exists.  
- Probability of Type I error is controlled by the significance level (α).  
- Example: A COVID-19 test incorrectly detects the virus in a healthy person.  

3. Understanding Type II Error:  
- Occurs when we fail to detect an effect when it actually exists.  
- Probability of Type II error is represented by β.  
- Power of the test (1 - β) measures the ability to detect a true effect.  
- Example: A COVID-19 test fails to detect the virus in an infected person.  

4. Comparison Table:  

| Error Type  | Meaning | Probability | Example |
|-------------|-----------------------------------|--------------|----------------------------------------|
| Type I Error (False Positive)  | Rejecting H0 when it is true  | α | Detecting a disease when the patient is healthy |
| Type II Error (False Negative) | Failing to reject H0 when it is false | β | Failing to detect a disease in an infected patient |

Q8. What is the Z-test, and when is it used in hypothesis testing?

1. What is a Z-Test?  
A **Z-test** is a statistical test used to determine if there is a significant difference between a sample mean and a population mean when the population variance is known or the sample size is large (n ≥ 30).  

2. When to Use a Z-Test?  
- The data follows a **normal distribution** (or n is large, Central Limit Theorem applies).  
- The **population standard deviation (σ) is known**.  
- The sample size is **large (n ≥ 30)**.  

3. Types of Z-Tests:  

| Z-Test Type | Null Hypothesis (H0) | Alternative Hypothesis (H1) | Use Case |
|------------|---------------------------------|---------------------------------|--------------------------------|
| **One-Sample Z-Test** | μ = μ₀ | μ ≠ μ₀ (or μ > μ₀, μ < μ₀) | Checking if the average height of students differs from 170 cm |
| **Two-Sample Z-Test** | μ₁ = μ₂ | μ₁ ≠ μ₂ (or μ₁ > μ₂, μ₁ < μ₂) | Comparing mean salaries of two different cities |

Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

1. What is a Z-Score?  
A **Z-score** (also known as a standard score) measures how many standard deviations a data point (or sample mean) is from the population mean. It helps determine how unusual or significant a result is in hypothesis testing.  

2. Formula for Z-Score:  

- For a single data point:  
\[
Z = \frac{X - \mu}{\sigma}
\]
Where:  
- **X** = observed value  
- **μ** = population mean  
- **σ** = population standard deviation  

- For a sample mean:  
\[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
\]
Where:  
- **X̄** = sample mean  
- **μ** = population mean  
- **σ** = population standard deviation  
- **n** = sample size  

3. Interpretation of Z-Score:  

| Z-Score | Interpretation |
|---------|---------------|
| Z = 0   | Value is exactly at the mean |
| Z = ±1  | Value is 1 standard deviation from the mean |
| Z = ±2  | Value is 2 standard deviations from the mean |
| Z = ±3  | Value is 3 standard deviations from the mean (rare event) |

It is similar to the normal distribution but has **heavier tails**, meaning it accounts for more variability in small samples.  

2. When to Use the T-Distribution Instead of the Normal Distribution?

|-----------|----------------|----------------|
| Sample size (n) | Large (≥ 30) | Small (< 30) |
| Population standard deviation (σ) | Known | Unknown |
| Data distribution | Normally distributed | Normally distributed |

3. Key Features of the T-Distribution:  
- **Similar to the normal distribution** but has heavier tails.  
- As **sample size (n) increases**, the t-distribution **approaches** the normal distribution.  
- Defined by **degrees of freedom (df = n - 1)**.  
- Used in **t-tests** for small-sample hypothesis testing.

Q12.  What is the T-test, and how is it used in hypothesis testing?

1. What is a T-Test?  
A **T-test** is a statistical test used to determine if there is a significant difference between the means of two groups, especially when the **sample size is small (n < 30)** and the **population standard deviation (σ) is unknown**. It relies on the **T-distribution**, which has heavier tails than the normal distribution.  

2. Types of T-Tests:  

| Type | Description | Use Case |
|------|------------|----------|
| **One-Sample T-Test** | Compares the sample mean to a known population mean | Checking if a sample differs from a population |
| **Two-Sample T-Test** | Compares the means of two independent groups | Testing differences between two groups (e.g., A/B testing) |
| **Paired T-Test** | Compares the means of the same group before and after treatment | Pre-test vs. Post-test analysis |

3. Steps for Performing a T-Test:  
   - Define the **null hypothesis (H₀)** and **alternative hypothesis (H₁)**.  
   - Calculate the **T-score** using the sample mean and sample standard deviation.  
   - Compute the **p-value** using the T-distribution.  
   - Compare the **p-value** with the significance level (α).  
   - If **p-value ≤ α**, reject the null hypothesis.

Q13. What is the relationship between Z-test and T-test in hypothesis testing?

1. **Overview**  
Both **Z-test** and **T-test** are used to determine if there is a significant difference between means in hypothesis testing. The key difference is that a **Z-test** is used when the **population standard deviation (σ) is known** or the **sample size is large (n ≥ 30)**, while a **T-test** is used when **σ is unknown** and the **sample size is small (n < 30)**.

2. **Similarities Between Z-Test and T-Test**  
- Both tests check whether a sample mean differs significantly from a population mean or another sample mean.  
- Both tests calculate a test statistic (**Z-score or T-score**) to determine statistical significance.  
- Both tests use a **p-value** to decide whether to reject the null hypothesis (**H₀**).  
- Both assume that the data follows a **normal distribution** (or approximately normal for large samples).  

3. **Differences Between Z-Test and T-Test**  

| Feature | Z-Test | T-Test |
|---------|-------|--------|
| **Use Case** | Large samples (n ≥ 30) | Small samples (n < 30) |
| **Population Standard Deviation (σ)** | Known | Unknown |
| **Distribution Used** | Normal Distribution (Z) | T-Distribution |
| **Tail Behavior** | Less tail area (less variability) | Heavier tails (more variability) |
| **Application** | Quality control, large population testing | Small-sample hypothesis testing |

4. **Mathematical Relationship**  
The formula for both tests is similar:  

- **Z-Test Formula:**  
  \[
  Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
  \]  
- **T-Test Formula:**  
  \[
  T = \frac{\bar{X} - \mu}{S / \sqrt{n}}
  \]  
  Where:  
  - \( \bar{X} \) = Sample mean  
  - \( \mu \) = Population mean  
  - \( \sigma \) = Population standard deviation (for Z-test)  
  - \( S \) = Sample standard deviation (for T-test)  
  - \( n \) = Sample size  

5. **When Do They Give Similar Results?**  
- When **n is large (n ≥ 30)**, the **T-distribution approaches the normal distribution**, making T-tests and Z-tests give almost identical results.  
- For **very large samples**, using the **Z-test** is preferred because the normal distribution is well-defined and computationally efficient.

Q14. What is a confidence interval, and how is it used to interpret statistical results?

1. **What is a Confidence Interval (CI)?**  
A **confidence interval (CI)** is a range of values that is likely to contain the true population parameter (such as mean or proportion) with a certain level of confidence. It helps in estimating the uncertainty in statistical results.  

2. **Why is a Confidence Interval Important?**  
- It provides a range instead of a single point estimate.  
- It helps in understanding the precision of the estimate.  
- It allows researchers to make inferences about a population from sample data.  

3. **Confidence Level (Common Values: 90%, 95%, 99%)**  
The confidence level represents the probability that the interval contains the true population parameter.  

- **95% CI**: There is a 95% probability that the true parameter lies within this interval.  
- **99% CI**: A wider interval that gives more certainty but less precision.  

4. **Formula for Confidence Interval**  

- For **population mean (μ)** when population standard deviation (σ) is known:  
\[
CI = \bar{X} \pm Z \times \left(\frac{\sigma}{\sqrt{n}}\right)
\]  

- For **small samples (n < 30) or unknown σ**, use the **T-distribution**:  
\[
CI = \bar{X} \pm T \times \left(\frac{S}{\sqrt{n}}\right)
\]  
Where:  
- \( \bar{X} \) = Sample mean  
- \( Z \) or \( T \) = Critical value from Z or T-table  
- \( \sigma \) = Population standard deviation (if known)  
- \( S \) = Sample standard deviation (if σ is unknown)  
- \( n \) = Sample size

Q15. What is the margin of error, and how does it affect the confidence interval?

1. **What is the Margin of Error (MoE)?**  
The **margin of error (MoE)** is the amount of uncertainty in an estimate. It tells us how much the sample estimate (like a mean or proportion) could vary from the true population value.  

2. **Why is the Margin of Error Important?**  
- It quantifies the **precision** of a confidence interval.  
- A **smaller MoE** means a **more precise** estimate.  
- A **larger MoE** means **more uncertainty** in the estimate.  

3. **Formula for Margin of Error**  

- For **large samples (Z-distribution, known σ)**:  
\[
MoE = Z \times \left(\frac{\sigma}{\sqrt{n}}\right)
\]  

- For **small samples (T-distribution, unknown σ)**:  
\[
MoE = T \times \left(\frac{S}{\sqrt{n}}\right)
\]  
Where:  
- \( Z \) or \( T \) = Critical value (from Z-table or T-table)  
- \( \sigma \) = Population standard deviation (if known)  
- \( S \) = Sample standard deviation (if σ is unknown)  
- \( n \) = Sample size  

4. **How Does MoE Affect the Confidence Interval?**  
The confidence interval (CI) is calculated as:  

- \[
CI = \bar{X} \pm MoE
\]  
Where \( \bar{X} \) is the sample mean.  

- **A larger MoE** → **Wider confidence interval**, meaning more uncertainty.  
- **A smaller MoE** → **Narrower confidence interval**, meaning a more precise estimate.

Q16. How is Bayes' Theorem used in statistics, and what is its significance?

1. **What is Bayes' Theorem?**  
Bayes' Theorem is a fundamental theorem in probability theory that describes how to update the probability of an event based on new evidence. It is widely used in **statistics, machine learning, medical diagnosis, spam filtering, and risk analysis**.  

2. **Formula for Bayes' Theorem**  

- \[
P(A | B) = \frac{P(B | A) \times P(A)}{P(B)}
\]

Where:  
- \( P(A | B) \) = Probability of event **A** occurring given that **B** has occurred (**Posterior probability**).  
- \( P(B | A) \) = Probability of event **B** occurring given that **A** is true (**Likelihood**).  
- \( P(A) \) = Prior probability of event **A**.  
- \( P(B) \) = Total probability of event **B**.  

3. **Why is Bayes' Theorem Important?**  
- It helps **update probabilities** when new data is available.  
- It is used in **classification problems** (e.g., Naïve Bayes classifier in Machine Learning).  
- It is useful in **medical testing** (e.g., probability of having a disease given a positive test result).

Q17.  What is the Chi-square distribution, and when is it used?

1. **What is the Chi-square Distribution?**  
The **Chi-square (χ²) distribution** is a probability distribution that arises in **hypothesis testing and variance estimation**. It is used in tests involving categorical data and population variance comparisons.  

2. **Formula for Chi-square Statistic**  

- \[
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
\]

Where:  
- \( O_i \) = Observed frequency  
- \( E_i \) = Expected frequency  

3. **Key Properties of Chi-square Distribution**  
- It is **skewed right** (asymmetrical).  
- It has **only positive values** (χ² ≥ 0).  
- The shape depends on **degrees of freedom (df)**.  

4. **When is the Chi-square Distribution Used?**  
- **Goodness-of-Fit Test**: Determines if a sample distribution fits an expected distribution.  
- **Independence Test**: Checks if two categorical variables are independent (e.g., gender vs. voting preference).  
- **Variance Testing**: Compares population variance with a theoretical value.

Q19. What is the F-distribution, and when is it used in hypothesis testing?

1. **What is the F-Distribution?**  
The **F-distribution** is a probability distribution used in **variance comparison** tests, such as the **ANOVA (Analysis of Variance) test** and the **F-test** for equality of variances.  
- It is a **right-skewed** distribution.  
- It is always **positive** (F-values ≥ 0).  
- It depends on two degrees of freedom: \( df_1 \) (numerator) and \( df_2 \) (denominator).  

2. **Formula for the F-statistic**  

- \[
F = \frac{S_1^2}{S_2^2}
\]

Where:  
- \( S_1^2 \) = Variance of sample 1 (larger variance).  
- \( S_2^2 \) = Variance of sample 2 (smaller variance).  

3. **When is the F-Distribution Used?**  
- **F-test**: Compares **two population variances**.  
- **ANOVA test**: Compares **multiple means** (to test if groups have different variances).  
- **Regression analysis**: Determines the **overall significance** of a regression model.

Q20. What is an ANOVA test, and what are its assumptions?

ANOVA (**Analysis of Variance**) is a statistical test used to compare the **means of three or more groups** to determine if there is a significant difference between them.  

- It uses the **F-distribution** to compare group variances.  
- It checks if at least **one group mean is different** from the others.  
- If differences exist, **post-hoc tests** (like Tukey’s test) determine which groups differ.

|------|------------|----------|

| **Two-Way ANOVA** | Compares means of **two factors** simultaneously | Studying the effect of **diet and exercise** on weight loss |
| **Repeated Measures ANOVA** | Compares means from the **same group over time** | Measuring blood pressure **before, during, and after** medication |

3. **ANOVA Hypotheses**  
- **Null Hypothesis (H₀):** All group means are equal.  
- **Alternative Hypothesis (H₁):** At least one group mean is different.  

4. **Assumptions of ANOVA**  

| Assumption | Description |
|------------|-------------|
| **Independence** | Samples must be **independent** of each other. |
| **Normality** | Data in each group should be **normally distributed**. |
| **Homogeneity of Variance (Homoscedasticity)** | All groups should have **equal variances** (checked using Levene’s test). |

```python
#Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test to compare the sample mean to a known population mean.

    Parameters:
    sample (list or numpy array): Sample data.
    population_mean (float): Known population mean.
    population_std (float): Known population standard deviation.
    alpha (float): Significance level (default is 0.05 for a 95% confidence level).

    Returns:
    tuple: Z-score, p-value, and interpretation of results.
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute the standard error
    standard_error = population_std / np.sqrt(sample_size)

    # Compute the Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Compute the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Interpretation
    if p_value < alpha:
        result = "Reject the null hypothesis: The sample mean is significantly different from the population mean."
    else:
        result = "Fail to reject the null hypothesis: No significant difference between the sample and population means."

    return z_score, p_value, result

# Example Usage
sample_data = [52, 55, 53, 54, 50, 49, 51, 55, 56, 52]  # Sample dataset
population_mean = 50  # Known population mean
population_std = 4    # Known population standard deviation

z_score, p_value, interpretation = z_test(sample_data, population_mean, population_std)
print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Interpretation: {interpretation}")
```

```python
#Q6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample, population_mean, population_std, alpha=0.05, tail='two'):
    """
    Perform a one-sample Z-test to compare the sample mean to a known population mean.
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    standard_error = population_std / np.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / standard_error
    if tail == 'two':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    elif tail == 'left':
        p_value = stats.norm.cdf(z_score)
    elif tail == 'right':
        p_value = 1 - stats.norm.cdf(z_score)
    else:
        raise ValueError("Invalid tail argument. Use 'two', 'left', or 'right'.")
    result = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"
    return z_score, p_value, result

def independent_t_test(sample1, sample2, alpha=0.05):
    """Perform an independent two-sample T-test and interpret the results."""
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)
    result = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"
    return t_stat, p_value, result

# Simulating random sample data
np.random.seed(42)
sample1 = np.random.normal(loc=52, scale=4, size=30)
sample2 = np.random.normal(loc=50, scale=4, size=30)

# Performing independent T-test
t_stat, p_value, interpretation = independent_t_test(sample1, sample2)
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Interpretation: {interpretation}")
```

```python
#Q16. Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two independent samples.

    Parameters:
    sample1 (array-like): First sample data.
    sample2 (array-like): Second sample data.
    alpha (float): Significance level (default is 0.05).

    Returns:
    f_stat (float): F-statistic (variance ratio).
    p_value (float): P-value from the F-distribution.
    result (str): Interpretation of the test result.
    """
    var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased)
    var2 = np.var(sample2, ddof=1)  # Sample variance (unbiased)

    # Compute F-statistic (larger variance in numerator)
    f_stat = var1 / var2 if var1 > var2 else var2 / var1

    # Degrees of freedom
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Compute one-tailed p-value
    p_value = 1 - stats.f.cdf(f_stat, df1, df2)

    # Interpret the result
    result = "Reject the null hypothesis (Variances are significantly different)" if p_value < alpha else "Fail to reject the null hypothesis (No significant difference in variances)"

    return f_stat, p_value, result

# Generate random samples (Normally distributed)
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean 50, Std 10, Size 30
sample2 = np.random.normal(loc=52, scale=15, size=30)  # Mean 52, Std 15, Size 30

# Perform F-Test
f_statistic, p_value, test_result = f_test(sample1, sample2)

# Display the results
print(f"F-Statistic: {f_statistic:.3f}")
print(f"P-Value: {p_value:.3f}")
print(f"Test Interpretation: {test_result}")
```

```python
#Q17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

import numpy as np
import scipy.stats as stats

def anova_test(*groups, alpha=0.05):
    """
    Perform a One-Way ANOVA test to compare means between multiple groups.

    Parameters:
    groups (list of array-like): Multiple groups of data.
    alpha (float): Significance level (default: 0.05).

    Returns:
    f_stat (float): F-statistic value.
    p_value (float): P-value from the ANOVA test.
    result (str): Interpretation of the test result.
    """
    f_stat, p_value = stats.f_oneway(*groups)

    # Interpretation of the test
    result = "Reject the null hypothesis (At least one group mean is significantly different)" if p_value < alpha else "Fail to reject the null hypothesis (No significant difference between means)"

    return f_stat, p_value, result

# Generate random data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10

# Perform ANOVA test
f_statistic, p_value, test_result = anova_test(group1, group2, group3)

# Display results
print(f"F-Statistic: {f_statistic:.3f}")
print(f"P-Value: {p_value:.3f}")
print(f"Test Interpretation: {test_result}")
```

```python
#Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import scipy.stats as stats

def check_anova_assumptions(*groups, alpha=0.05):
    """
    Check assumptions of ANOVA: Normality, Independence, and Homogeneity of Variance.

    Parameters:
    groups (list of array-like): Multiple groups of data.
    alpha (float): Significance level for tests (default: 0.05).

    Returns:
    dict: Results of normality and variance tests.
    """
    results = {}

    # 1. Normality Check (Shapiro-Wilk Test)
    normality_results = []
    for i, group in enumerate(groups, 1):
        stat, p_value = stats.shapiro(group)
        normality_results.append(f"Group {i}: {'Pass' if p_value > alpha else 'Fail'} (p={p_value:.3f})")

    results['Normality'] = normality_results

    # 2. Homogeneity of Variance (Levene’s Test)
    levene_stat, levene_p = stats.levene(*groups)
    results['Equal Variance'] = f"{'Pass' if levene_p > alpha else 'Fail'} (p={levene_p:.3f})"

    return results

# Generate sample data
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

# Check ANOVA assumptions
assumption_results = check_anova_assumptions(group1, group2, group3)

# Print results
print("ANOVA Assumption Check:")
for key, value in assumption_results.items():
    print(f"{key}: {value}")
```

```python
#Q21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_f_distribution(df1_values, df2, xlim=(0, 5)):
    """
    Plot the F-distribution for different numerator degrees of freedom (df1).

    Parameters:
    df1_values (list): List of numerator degrees of freedom (df1) to plot.
    df2 (int): Denominator degrees of freedom.
    xlim (tuple): Range of x-axis.
    """
    x = np.linspace(xlim[0], xlim[1], 1000)  # Generate x values
    plt.figure(figsize=(8, 6))

    for df1 in df1_values:
        y = stats.f.pdf(x, df1, df2)  # F-distribution PDF
        plt.plot(x, y, label=f"df1={df1}, df2={df2}")

    plt.title("F-Distribution for Different df1 Values")
    plt.xlabel("F-Value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage with different numerator degrees of freedom
plot_f_distribution(df1_values=[1, 5, 10], df2=20)
```

```python
#Q22.  Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Simulate sample data
np.random.seed(42)
group_1 = np.random.normal(loc=50, scale=5, size=30)  # Group A
group_2 = np.random.normal(loc=55, scale=5, size=30)  # Group B
group_3 = np.random.normal(loc=60, scale=5, size=30)  # Group C

# Combine data into a DataFrame
df = pd.DataFrame({
    "Score": np.concatenate([group_1, group_2, group_3]),
    "Group": np.repeat(["A", "B", "C"], 30)
})

# Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(group_1, group_2, group_3)

# Display ANOVA results
print(f"F-Statistic: {f_stat:.3f}")
print(f"P-Value: {p_value:.3f}")

# Interpret results
if p_value < 0.05:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")

# Visualization - Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x="Group", y="Score", data=df, palette="Set2")
plt.title("One-Way ANOVA: Group Comparisons")
plt.xlabel("Groups")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
#Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate Random Data from a Normal Distribution
np.random.seed(42)
sample_size = 30  # Number of observations
population_mean = 50
population_std = 5  # Known population standard deviation

# Generate a random sample
sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Perform a One-Sample T-Test (when population variance is unknown)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Sample standard deviation
t_stat, p_value = stats.ttest_1samp(sample, popmean=population_mean)

# Step 3: Perform a One-Sample Z-Test (when population variance is known)
standard_error = population_std / np.sqrt(sample_size)
z_score = (sample_mean - population_mean) / standard_error
z_p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"T-Test: T-Statistic = {t_stat:.3f}, P-Value = {p_value:.3f}")
print(f"Z-Test: Z-Score = {z_score:.3f}, P-Value = {z_p_value:.3f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("T-Test: Reject the null hypothesis (significant difference in means).")
else:
    print("T-Test: Fail to reject the null hypothesis (no significant difference).")

if z_p_value < alpha:
    print("Z-Test: Reject the null hypothesis (significant difference in means).")
else:
    print("Z-Test: Fail to reject the null hypothesis (no significant difference).")

# Step 4: Visualize the Sample Distribution
plt.figure(figsize=(8, 5))
plt.hist(sample, bins=10, alpha=0.7, color='skyblue', edgecolor='black', density=True, label="Sample Data")
x = np.linspace(min(sample), max(sample), 100)
plt.plot(x, stats.norm.pdf(x, loc=sample_mean, scale=sample_std), color='red', label="Fitted Normal Curve")
plt.axvline(population_mean, color='black', linestyle="--", label="Population Mean")
plt.axvline(sample_mean, color='green', linestyle="--", label="Sample Mean")
plt.title("Sample Distribution with Normal Fit")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
import scipy.stats as stats

def chi_square_variance_test(sample, population_variance, alpha=0.05, tail='two'):
    """
    Perform a Chi-square test for population variance.

    Parameters:
    - sample: The dataset (array-like)
    - population_variance: Hypothesized population variance (σ₀²)
    - alpha: Significance level (default = 0.05)
    - tail: Type of test ('two', 'left', or 'right')

    Returns:
    - Chi-square statistic, p-value, decision (reject/fail to reject H₀)
    """
    sample_size = len(sample)
    sample_variance = np.var(sample, ddof=1)  # Sample variance (unbiased)

    # Compute Chi-square test statistic
    chi_square_stat = (sample_size - 1) * sample_variance / population_variance

    # Determine critical values and p-value
    if tail == 'two':
        lower_critical = stats.chi2.ppf(alpha / 2, df=sample_size - 1)
        upper_critical = stats.chi2.ppf(1 - alpha / 2, df=sample_size - 1)
        p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df=sample_size - 1),
                          1 - stats.chi2.cdf(chi_square_stat, df=sample_size - 1))
        reject_h0 = chi_square_stat < lower_critical or chi_square_stat > upper_critical
    elif tail == 'left':
        critical_value = stats.chi2.ppf(alpha, df=sample_size - 1)
        p_value = stats.chi2.cdf(chi_square_stat, df=sample_size - 1)
        reject_h0 = chi_square_stat < critical_value
    elif tail == 'right':
        critical_value = stats.chi2.ppf(1 - alpha, df=sample_size - 1)
        p_value = 1 - stats.chi2.cdf(chi_square_stat, df=sample_size - 1)
        reject_h0 = chi_square_stat > critical_value
    else:
        raise ValueError("Invalid tail argument. Use 'two', 'left', or 'right'.")

    # Decision
    decision = "Reject the null hypothesis" if reject_h0 else "Fail to reject the null hypothesis"

    return chi_square_stat, p_value, decision

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=30)  # Sample with known population variance
hypothesized_variance = 25  # Population variance (σ₀² = 5²)

chi2_stat, p_value, decision = chi_square_variance_test(sample_data, hypothesized_variance, alpha=0.05, tail='two')

# Output Results
print(f"Chi-Square Statistic: {chi2_stat:.3f}")
print(f"P-Value: {p_value:.3f}")
print(f"Decision: {decision}")
```

```python
#Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
import scipy.stats as stats

def z_test_proportions(success1, n1, success2, n2, alpha=0.05, tail='two'):
    """
    Perform a Z-test for comparing proportions between two groups.

    Parameters:
    - success1: Number of successes in group 1
    - n1: Sample size of group 1
    - success2: Number of successes in group 2
    - n2: Sample size of group 2
    - alpha: Significance level (default = 0.05)
    - tail: Type of test ('two', 'left', 'right')

    Returns:
    - Z-score, p-value, and decision (reject/fail to reject H₀)
    """
    # Compute sample proportions
    p1 = success1 / n1
    p2 = success2 / n2

    # Compute pooled proportion
    p_pool = (success1 + success2) / (n1 + n2)

    # Compute standard error
    standard_error = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

    # Compute Z-score
    z_score = (p1 - p2) / standard_error

    # Compute p-value based on the type of test
    if tail == 'two':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    elif tail == 'left':
        p_value = stats.norm.cdf(z_score)
    elif tail == 'right':
        p_value = 1 - stats.norm.cdf(z_score)
    else:
        raise ValueError("Invalid tail argument. Use 'two', 'left', or 'right'.")

    # Decision
    decision = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

    return z_score, p_value, decision

# Example Usage
success_group1 = 45  # Number of successes in group 1
total_group1 = 100   # Total sample size in group 1

success_group2 = 30  # Number of successes in group 2
total_group2 = 100   # Total sample size in group 2

# Perform the Z-test for proportions
z_stat, p_val, decision = z_test_proportions(success_group1, total_group1, success_group2, total_group2, alpha=0.05, tail='two')

# Print results
print(f"Z-Score: {z_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(f"Decision: {decision}")
```

```python
#Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test_variance(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two samples.
    """
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        F_stat = var1 / var2
        df1, df2 = len(sample1) - 1, len(sample2) - 1
    else:
        F_stat = var2 / var1
        df1, df2 = len(sample2) - 1, len(sample1) - 1

    p_value = 2 * min(stats.f.cdf(F_stat, df1, df2), 1 - stats.f.cdf(F_stat, df1, df2))
    decision = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

    return F_stat, p_value, decision, df1, df2

# Generate random samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

# Perform the F-test
F_stat, p_val, decision, df1, df2 = f_test_variance(sample1, sample2)

# Print results
print(f"F-Statistic: {F_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(f"Decision: {decision}")

# Visualization: F-distribution and critical region
x = np.linspace(0, 5, 1000)
y = stats.f.pdf(x, df1, df2)  # ✅ Fixed here

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='F-Distribution', color='blue')
plt.axvline(F_stat, color='red', linestyle='--', label=f'Observed F = {F_stat:.3f}')
plt.fill_between(x, y, where=(x >= stats.f.ppf(1-alpha/2, df1, df2)), color='red', alpha=0.3, label=f'α = {alpha} (Critical Region)')
plt.fill_between(x, y, where=(x <= stats.f.ppf(alpha/2, df1, df2)), color='red', alpha=0.3)

plt.xlabel('F-Value')
plt.ylabel('Probability Density')
plt.title('F-Test for Variance Comparison')
plt.legend()
plt.show()
```

```python
#Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def chi_square_goodness_of_fit(observed, expected, alpha=0.05):
    """
    Perform a Chi-square goodness-of-fit test.

    Parameters:
    - observed: List/array of observed frequencies
    - expected: List/array of expected frequencies
    - alpha: Significance level (default = 0.05)

    Returns:
    - Chi-square statistic, p-value, and interpretation
    """
    # Ensure expected values sum up to observed values
    expected = expected * (np.sum(observed) / np.sum(expected))

    chi2_stat, p_value = stats.chisquare(observed, expected)

    # Decision rule
    decision = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

    return chi2_stat, p_value, decision

# Simulated observed data
np.random.seed(42)
observed_counts = np.random.randint(10, 50, size=5)  # 5 categories

# Expected frequencies (normalized to match observed total)
expected_counts = np.full_like(observed_counts, np.mean(observed_counts))

# Perform Chi-square goodness-of-fit test
chi2_stat, p_val, decision = chi_square_goodness_of_fit(observed_counts, expected_counts)

# Print results
print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(f"Decision: {decision}")

# Visualization of observed vs. expected counts
categories = [f'Cat {i+1}' for i in range(len(observed_counts))]

plt.figure(figsize=(8, 5))
plt.bar(categories, observed_counts, alpha=0.7, label='Observed', color='blue')
plt.bar(categories, expected_counts, alpha=0.5, label='Expected', color='orange')
plt.xlabel("Categories")
plt.ylabel("Frequency")
plt.title("Observed vs. Expected Frequencies")
plt.legend()
plt.show()

```
